Персонажи
Shrek, PussInBoots, Donkey, JackHorner
на каждого должен быть и персонаж, и фанко поп, и косплеер
итого должно быть:
BaseCharacter: Shrek PussInBoots Donkey JackHorner
BaseCharacter -> BaseInActionCharacter: Shrek -> ShrekInAction PussInBoots -> PussInBootsInAction Donkey -> DonkeyInAction JackHorner -> JackHornerInAction
BaseCharacter -> BaseFunkoPop: Shrek -> ShrekInAction PussInBoots -> PussInBootsInAction Donkey -> DonkeyInAction JackHorner -> JackHornerInAction
BaseCharacter -> BaseCosplayer <- BaseHuman: Shrek -> ShrekCosplayer PussInBoots -> PussInBootsCosplayer Donkey -> DonkeyCosplayer JackHorner -> JackHornerCosplayer

```python
class MixinSpeakable:
    def speak(self):
        return f"{self.name} говорит чтото странное"

class MixinActionable:
    def perform_action(self):
        return f"{self.name} делает чтото эпичное"

class MixinCollectible:
    def display(self):
        return f"Фанкопоп {self.name} смотрит в пустоту"

class MixinPoseable:
    def pose(self):
        return f"{self.name} замирает в дурацкой позе"

class MixinCostumeWearable:
    def __init__(self):
        self.costume = " супер странный костюм"

class MixinAnimated:
    def __init__(self):
        self.is_animated = True

class MixinFunny:
    def make_laugh(self):
        return f"{self.name} заставляет всех смеяться"

class BaseCharacter(MixinSpeakable, MixinAnimated, MixinFunny):
    def __init__(self, name):
        self.name = name

class BaseInActionCharacter(BaseCharacter, MixinActionable):
    pass

class BaseFunkoPop(BaseCharacter, MixinCollectible):
    pass

class BaseHuman:
    def __init__(self):
        self.is_human = True

class BaseCosplayer(BaseHuman, MixinPoseable, MixinCostumeWearable):
    def __init__(self, character_name):
        super().__init__()
        self.name = f"косплеер {character_name}"

class Shrek(BaseCharacter):
    def __init__(self):
        super().__init__("шрек")

class PussInBoots(BaseCharacter):
    def __init__(self):
        super().__init__("кот в сапогах")

class Donkey(BaseCharacter):
    def __init__(self):
        super().__init__("осел")

class JackHorner(BaseCharacter):
    def __init__(self):
        super().__init__("джек")

class ShrekInAction(BaseInActionCharacter):
    def __init__(self):
        super().__init__("шрек в действии")

class PussInBootsInAction(BaseInActionCharacter):
    def __init__(self):
        super().__init__("кот в сапогах в действии")

class DonkeyInAction(BaseInActionCharacter):
    def __init__(self):
        super().__init__("осел в действии")

class JackHornerInAction(BaseInActionCharacter):
    def __init__(self):
        super().__init__("джек в действии")

class ShrekFunkoPop(BaseFunkoPop):
    def __init__(self):
        super().__init__("шрек Фанкопоп")

class PussInBootsFunkoPop(BaseFunkoPop):
    def __init__(self):
        super().__init__("кот в сапогах Фанкопоп")

class DonkeyFunkoPop(BaseFunkoPop):
    def __init__(self):
        super().__init__("осел Фанкопоп")

class JackHornerFunkoPop(BaseFunkoPop):
    def __init__(self):
        super().__init__("джек Фанкопоп")

class ShrekCosplayer(BaseCosplayer):
    def __init__(self):
        super().__init__("на шрека")

class PussInBootsCosplayer(BaseCosplayer):
    def __init__(self):
        super().__init__("на кота в сапогах")

class DonkeyCosplayer(BaseCosplayer):
    def __init__(self):
        super().__init__("на осла")

class JackHornerCosplayer(BaseCosplayer):
    def __init__(self):
        super().__init__("на джека")

shrek = Shrek()
shrek_action = ShrekInAction()
shrek_funko = ShrekFunkoPop()
shrek_cosplay = ShrekCosplayer()

puss = PussInBoots()
puss_action = PussInBootsInAction()
puss_funko = PussInBootsFunkoPop()
puss_cosplay = PussInBootsCosplayer()

donkey = Donkey()
donkey_action = DonkeyInAction()
donkey_funko = DonkeyFunkoPop()
donkey_cosplay = DonkeyCosplayer()

jack = JackHorner()
jack_action = JackHornerInAction()
jack_funko = JackHornerFunkoPop()
jack_cosplay = JackHornerCosplayer()

print("шрек")
print(shrek.speak())
print(shrek.make_laugh())
print(shrek_action.perform_action())
print(shrek_funko.display())
print(shrek_cosplay.pose())

print("пуси в сапогах лол")
print(puss.speak())
print(puss.make_laugh())
print(puss_action.perform_action())
print(puss_funko.display())
print(puss_cosplay.pose())

print("осел")
print(donkey.speak())
print(donkey.make_laugh())
print(donkey_action.perform_action())
print(donkey_funko.display())
print(donkey_cosplay.pose())

print("джек")
print(jack.speak())
print(jack.make_laugh())
print(jack_action.perform_action())
print(jack_funko.display())
print(jack_cosplay.pose())
```
