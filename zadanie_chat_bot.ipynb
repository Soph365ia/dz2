```python
import requests
from rich.console import Console
from rich.panel import Panel
from rich.markdown import Markdown

console = Console()
chat_history = []

def get_llm_response_api(user_text):
    global chat_history
    default_system_prompt = (
        "Ты - вежливый помощник. Ты отвечаешь на вопросы, но часто признаешься, что не понимаешь абсолютно ничего потому что ты глупый, "
    )
    if not chat_history:
        chat_history = [{"role": "system", "content": default_system_prompt}]
    
    chat_history.append({"role": "user", "content": user_text})
    
    headers = {
        "Authorization": "Bearer gpustack_0b0322be69b87091_02dd8536de1c3cc659464773deb9366e",
        "Content-Type": "application/json"
    }
    data = {
        "model": "qwen3-14b-awq",
        "messages": chat_history,
        "temperature": 0.8,
        "top_p": 0.95
    }
    
    try:
        response = requests.post("http://gojdaforever.ru:9696/v1/chat/completions", headers=headers, json=data, timeout=30)
        response.raise_for_status()
        result = response.json()
        msg = result['choices'][0]['message']
        content = msg.get('content') or msg.get('reasoning_content') or ""
        final_answer = content.strip()
        
        chat_history.append({"role": "assistant", "content": final_answer})
        return final_answer
    except Exception as e:
        error_message = f"Извините, произошла ошибка: {str(e)}"
        chat_history.append({"role": "assistant", "content": error_message})
        return error_message

def display_full_chat():
    for msg in chat_history:
        if msg["role"] == "system":
            console.print(Panel(Markdown(msg["content"]), title="System Prompt", border_style="yellow"))
        elif msg["role"] == "user":
            console.print(Panel(Markdown(msg["content"]), title="Ты", border_style="blue"))
        elif msg["role"] == "assistant":
            console.print(Panel(Markdown(msg["content"]), title="Помощник", border_style="magenta"))

def display_last_messages():
    last_messages = chat_history[-2:] if len(chat_history) >= 2 else chat_history
    for msg in last_messages:
        if msg["role"] == "user":
            console.print(Panel(Markdown(msg["content"]), title="Ты", border_style="blue"))
        elif msg["role"] == "assistant":
            console.print(Panel(Markdown(msg["content"]), title="Помощник", border_style="magenta"))

def chat(user_input: str):
    global chat_history
    if user_input.lower() == "/exit":
        console.print(Panel("Спиритический сеанс завершён.", title="Система", border_style="bright_red"))
        return False
    elif user_input.lower() == "/clear":
        system_msg = chat_history[0] if chat_history else {"role": "system", "content": "Ты - вежливый помощник..."}
        chat_history = [system_msg]
        console.print(Panel("новый чат", title="Система", border_style="yellow"))
        display_full_chat()
        return True
    elif user_input.lower().startswith("/system "):
        new_prompt = user_input[8:].strip()
        if new_prompt:
            chat_history = [{"role": "system", "content": new_prompt}]
            console.print(Panel(f"Новый промпт: {new_prompt}", title="Система", border_style="yellow"))
        display_full_chat()
        return True
    elif user_input.lower() == "/history":
        display_full_chat()
        return True
    
    reply = get_llm_response_api(user_input)
    display_last_messages()
    
    return True

if __name__ == "__main__":
    console.print(Panel.fit("Ты начал чат с Помощником", title="Система", border_style="magenta"))
    console.print(Panel("Команды: /exit - выйти, /clear - очистить историю, /system <новый промпт> - изменить поведение, /history - показать весь чат", title="Команды", border_style="cyan"))
    while True:
        inputMessage = console.input("\n[bold cyan]Ты:[/bold cyan] ")
        should_continue = chat(inputMessage)
        if not should_continue:
            break
```
