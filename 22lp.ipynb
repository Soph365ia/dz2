```python
def chisla():
    a = -1000
    b = 1000
    
    while a < b:
        mid = (a + b) // 2
        otvet = input(f"Число меньше {mid}? ")
        
        if otvet == "да":
            b = mid - 1
        else:
            a = mid
            
        if a == b:
            print(f"твое супер число: {a}")
            return
    
    print(f"твое супер число: {a}")

chisla()
```
